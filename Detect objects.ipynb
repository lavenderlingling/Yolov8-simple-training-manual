## 捕捉動態照片

```python
import cv2
import time
from ultralytics import YOLO
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

# 加載YOLOv8模型
model = YOLO("C:/Users/kevin/Work/photo capture project/best.pt")

# 讀取CCD串流資料
vs = cv2.VideoCapture(0)
time.sleep(1.0)
flg = True

while flg:
    flg, frame = vs.read()
    if not flg :
        break

    # 在幀上運行YOLOv8追蹤，持續追蹤幀間的物體
    results = model.track(frame, persist=True)

    # 在幀上展示結果
    annotated_frame = results[0].plot() 
        
    # 展示帶上註釋的幀
    cv2.imshow("YOLOv8 Tracking", annotated_frame)
    
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break

# 釋放視頻捕獲對象並關閉顯示窗口
vs.release()
cv2.destroyAllWindows()
```
